```python
graph = {
    'A': [('B', 4), ('C', 3)],
    'B': [('A', 4), ('C', 1), ('D', 2)],
    'C': [('A', 3), ('B', 1), ('D', 4), ('E', 5)],
    'D': [('B', 2), ('C', 4), ('E', 1)],
    'E': [('C', 5), ('D', 1)]
}
```

```python
import heapq

def prim_mst(graph, start):
    visited = set()
    min_heap = [(0, start, None)]

    mst_cost = 0
    mst_edges = []

    while min_heap:
        cost, node, parent = heapq.heappop(min_heap)

        if node in visited:
            continue

        visited.add(node)
        mst_cost += cost

        if parent is not None:
            mst_edges.append((parent, node, cost))

        for neighbor, weight in graph[node]:
            if neighbor not in visited:
                heapq.heappush(min_heap, (weight, neighbor, node))

    return mst_cost, mst_edges
```

```python
graph = {
    'A': [('B', 4), ('C', 3)],
    'B': [('A', 4), ('C', 1), ('D', 2)],
    'C': [('A', 3), ('B', 1), ('D', 4), ('E', 5)],
    'D': [('B', 2), ('C', 4), ('E', 1)],
    'E': [('C', 5), ('D', 1)]
}

start_node = 'A'
```

```python
cost, edges = prim_mst(graph, start_node)
print("Minimum Cost:", cost)
print("Edges in MST:", edges)
```
